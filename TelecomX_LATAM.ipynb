```python
import pandas as pd

df = pd.read_json('/content/TelecomX_Data.json')

df = df[df['Churn'] != '']

customer_df = df['customer'].apply(pd.Series)
phone_df = df['phone'].apply(pd.Series)
internet_df = df['internet'].apply(pd.Series)
account_df = df['account'].apply(pd.Series)

df = pd.concat([df.drop(columns=['customer', 'phone', 'internet', 'account']), customer_df, phone_df, internet_df, account_df], axis=1)

df['TotalCharges'] = df['Charges'].apply(lambda x: x.get('Total') if isinstance(x, dict) else None)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)

df = df.drop(columns=['Charges'])

display(df.head())
display(df.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

display(df['Churn'].value_counts())

plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=df)
plt.title('Distribution of Churn')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()
```

```python
categorical_features = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']

for feature in categorical_features:
    churn_rate = df.groupby(feature)['Churn'].value_counts(normalize=True).unstack()
    display(f"Churn rate by {feature}:")
    display(churn_rate)
```

```python
numerical_features = ['tenure', 'TotalCharges']

for feature in numerical_features:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=feature, hue='Churn', multiple='stack', kde=True)
    plt.title(f'Distribution of {feature} by Churn')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

    display(f"Summary statistics for {feature} by Churn:")
    display(df.groupby('Churn')[feature].describe())
```

```python
print("Potential features with strong correlation with churn based on the analysis:")
print("- SeniorCitizen: Higher churn rate for senior citizens.")
print("- Partner: Customers with no partners have a higher churn rate.")
print("- Dependents: Customers without dependents have a higher churn rate.")
print("- InternetService: Fiber optic internet service has a significantly higher churn rate compared to DSL or no internet service.")
print("- OnlineSecurity: Customers without online security have a much higher churn rate.")
print("- OnlineBackup: Customers without online backup have a higher churn rate.")
print("- DeviceProtection: Customers without device protection have a higher churn rate.")
print("- TechSupport: Customers without technical support have a much higher churn rate.")
print("- Contract: Month-to-month contracts have a substantially higher churn rate than one-year or two-year contracts.")
print("- PaperlessBilling: Customers with paperless billing have a higher churn rate.")
print("- PaymentMethod: Electronic check payment method has the highest churn rate.")
print("- tenure: Customers who churn tend to have lower tenure.")
print("- TotalCharges: Customers who churn tend to have lower total charges.")
```
